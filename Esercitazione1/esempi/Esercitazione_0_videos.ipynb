Handling videos is generally very complex but openCV provides tools for making it easier.
A video is a sequence of images, called frames. 

The number of frames per second is defined frame rate of fps (frames per second). Typical value are 25 (in EU) or 30 (in US) but other values are common nowadays.

All frames must have all the same size. Typical resolution (WxH) are:
- 1280 x 720 (HD or 720p)
- 1920 x 1080 (Full HD or 1080p)
- 3840 x 2160 (4K or Ultra HD or 2160p)

Storing a 60 seconds color video in HD at 25 fps requires saving 1500 frames on HDD. 
Each frame occupies 1280x720x3 byte (about 2.8 MB). Storing this short video would require 4.2 GB.

However, many of these frames will look pretty similar. Thus, it is possible to use compression techniques to save space. Compression is also important for streaming since it reduces the bandwidth required to stream the video.
Moreover, to enhance compatibility among devices and platforms, compressed videos must be coded using suitable standard.
Video codec refers to compression standards. The video coding process requires a coding step (done through the encoder) to compress data and a decoding step (through a decoder) to approximate the original video (when compression determines a loss of information).

The most used codecs include: H.264 (especially for streaming, is also called AVC), H.265 (called also HEVC), VP9 and AV1.

Codec and video format are not the same.
Codec is the compression-decompression algorithm.
Video format indicates the type of file, which is like a container for the compressed video but also for other data, such as audio or subtitles. 
Common video formats are: MKV and MP4 (also known as MPEG 4).

Due to the compression algorithm, it is difficult to read a random frame from a video. Indeed, the compression of this frame may involve also information extracted from previous frames (depending on the compression algorithm).

The steps to handle a video are similar to those required when handling files:
- open the video;
- read frame-by-frame (for instance, within a for loop) till the end of the video;
- close the video (and free resources allocated).

In openCV, videos are handled by an object of the class VideoCapture. 
The video can be opened by the constructor of the object or through the method open(). Frames are read by the method read(). The method release() allows closing the file.




```python
import numpy as np
import cv2

video = cv2.VideoCapture("Tracking.mp4")
if video is None or not video.isOpened():
    raise IOError("video not found")
else:
    ret, img = video.read()
    while ret:
        cv2.imshow("frame", img)
        cv2.waitKey(1)
        ret, img = video.read()
    video.release()
        
        
```

We can also recover information about the video by using the method get, which takes in input a parameter (a constant). The list of parameters is at the link https://docs.opencv.org/3.4/d4/d15/group__videoio__flags__base.html#gaeb8dd9c89c10a5c63c139bf7c4f5704d

```python
video = cv2.VideoCapture("Tracking.mp4")

print("number of frames ", video.get(cv2.CAP_PROP_FRAME_COUNT  ))
print("CONVERT_RGB ", video.get(cv2.CAP_PROP_CONVERT_RGB ))
 
```

To write a new video, we need to use an object of the class VideoWriter.
Here, we need to define the codec and the format of the video. 
OpenCV uses the extension of the file to infer the format, while the codec can be specified explicitly by indicating a 4-byte code called FourCC (may change based on the OS).

The list of FourCC can be found at the link

Be aware that some codecs require the installation of external library to work. (Indeed, it is reccommended to install ffmpeg).


The code below 



```python
import numpy as np
import cv2

video = cv2.VideoCapture("Tracking.mp4")

video_FourCC    = int(video.get(cv2.CAP_PROP_FOURCC)).to_bytes(4,"little")
print("FourCC of video in input ", video_FourCC)

fps = video.get(cv2.CAP_PROP_FPS)
H = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT)/2)
W = int(video.get(cv2.CAP_PROP_FRAME_WIDTH)/2)
frame_size = (W, H)

#out = cv2.VideoWriter('rescaled.mp4',  cv2.VideoWriter.fourcc('m','p','4','v'), fps, frame_size)
out = cv2.VideoWriter('rescaled_h264.mp4',  cv2.VideoWriter.fourcc('a','v','c','1'), fps, frame_size)

if video is None or not video.isOpened() or out is None or not out.isOpened() :
    raise IOError("video not found")
else:
    ret, img = video.read()
    while ret:
        cv2.imshow("frame", img)
        cv2.waitKey(1)
        img2 = cv2.resize(img, (0,0), fx=0.5, fy=0.5)
        out.write(img2)
        ret, img = video.read()
    video.release()
    out.release()
    
```

A video is essentially a stream (as the ones you have studied in C).

If we want to read frames from a camera, we need to set up a stream as well. It is possible to use videoCapture to acquire a stream video from a camera.

We need to specify to the object VideoCapture the device we want to use. The method set will allow setting the camera parameters (changing frame size or fps....) depending on the API provided with the camera

OpenCV assigns a numerical ID between 0-99 to each of the camera directly connected to the computer.
If you just have a webcam, its ID will probably be 0. 

For IP cameras, you need to specify the url to stream the video. There are several streaming protocols. We will use rtsp.


```python
import numpy as np
import cv2


ID = 0 # webcam
video = cv2.VideoCapture(ID)
print(video)
if video is None or not video.isOpened():
    raise IOError("video not found")
else:
    print("frame width ", video.get(cv2.CAP_PROP_FRAME_WIDTH ))
    print("frame height ", video.get(cv2.CAP_PROP_FRAME_HEIGHT ))
    print("frame rate ", video.get(cv2.CAP_PROP_FPS ))
    ret, img = video.read()
    key = ''
    while ret and key!=ord('q'):
        cv2.imshow("frame", img)
        key = cv2.waitKey(1)
        ret, img = video.read()
        
    video.release()
```
